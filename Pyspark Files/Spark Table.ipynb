```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName('Table Demo') \
.enableHiveSupport() \
.getOrCreate()
```

```python
data = [
    (0,'Customer_0','Pune','Maharashtra','India','29-06-2023','FALSE'),
    (1,'Customer_1','Bangalore','Tamil Nadu','India','07-12-2023','TRUE'),
    (2,'Customer_2','Hyderabad','Gujarat','India','27-10-2023','TRUE'),
    (3,'Customer_3','Bangalore','Karnataka','India','17-10-2023','FALSE')]

columns = ['customer_id','name','city','state','country','registration_date','is_active']
```

```python
spark.sql('show tables').show() # To check if there are any tables in the default database
```

```python
spark.sql('drop table default.customers_persistent')
```

```python
df = spark.createDataFrame(data, schema=columns)
```

```python
df.createOrReplaceTempView('temp_customers')
```

```python
spark.sql('select * from temp_customers limit 2').show()
```

## Global Temporary Table

```python
df.createOrReplaceGlobalTempView('global_customers')
```

```python
spark.sql('show tables in global_temp').show()
```

```python
spark.sql('select * from global_temp.global_customers limit 2').show()
```

```python
spark.sql('drop table global_customers')
```

```python
spark.sql('show tables in global_temp').show()
```

## Persistent Table

```python
df.write.mode('overwrite').saveAsTable('customers_persistent')
```

```python
spark.sql('show tables').show()
```

```python
spark.sql('select * from customers_persistent limit 3').show()
```

```python
spark.sql('describe customers_persistent').show()
```

```python
spark.sql('describe extended customers_persistent').show(truncate = False)
```

```python
spark.stop()
```
