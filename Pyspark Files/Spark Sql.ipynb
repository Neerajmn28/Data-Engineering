```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
.appName('Spark SQL') \
.enableHiveSupport() \
.getOrCreate()
```

```python
# Data

data = [(1, 'Alice','Mumbai','2023-01-15',True),
        (2, 'Bob','Delhi','2023-03-25',False),
        (3,'Charlie','Chennai','2023-05-10',True)
]

columns = ['customer_id','name','city','registration_date','is_date']

df = spark.createDataFrame(data,columns)
```

```python
spark.sql('show databases').show()
```

```python
spark.sql('use default')
```

```python
spark.sql('show tables').show()
```

```python
spark.sql('select * from customers_persistent limit 5').show()
```

```python
spark.sql('''
CREATE TABLE IF NOT EXISTS customers (
customer_id INT,
name STRING,
city STRING,
registration_date STRING,
is_active BOOLEAN)
USING CSV
''')
```

```python
spark.sql('show tables').show()
```

```python
spark.sql('describe extended customers').show(truncate = False)
```

```python
df.write.mode('overwrite').saveAsTable('default.customers')
```

```python
spark.sql('select * from customers limit 5').show()
```

```python
spark.stop()
```
