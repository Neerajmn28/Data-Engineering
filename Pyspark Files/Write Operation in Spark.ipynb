```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
.appName('Write_in_Spark') \
.getOrCreate()
```

```python
data = [(1,'Alice','Mumbai','2023-01-15',True),
       (2,'Bob','Delhi','2023-03-25', False),
       (3,'Charlie','Chennai','2023-05-10',True)]
```

```python
columns = ['customer_id','name','city','registration_date','is_active']
```

```python
df = spark.createDataFrame(data,columns)
```

```python
df.write\
.format('csv') \
.option('header','true') \
.save('/data/write_output.csv')
```

```python
df.rdd.getNumPartitions()
```

```python
df.repartition(1).write \
.format('csv') \
.option('header','true') \
.option('delimeter','|') \
.save('/data/write_output_2_part')
```
