```python
from pyspark.sql import SparkSession


# Initialize Spark Session
spark = SparkSession.builder \
.appName('Data Frame_caching_demo') \
.getOrCreate()
```

```python
customers_df = spark.read.option('header','true').csv('/user/customers.csv')
```

```python
customers_df.printSchema()
```

```python
customers_df.show(5)
```

```python
customers_df.cache()
```

```python
filtered_df = customers_df.filter(customers_df.city == 'Mumbai')
filtered_df.count()
```

```python
customers_df.unpersist()
```

```python
spark.stop()
```
